###FILTER/WHERE

###scenario-1 
####filter the data with fat content = regular
####slice the data with item type = soft drinks and weight<10
####fetch the data with Tier in (tier 1 or tier 2) and outlet size is null

```python
from pyspark.sql.functions import col
df.filter(col('Item_Fat_Content')=='Regular').display()
```

###scenario-2


```python
from pyspark.sql.functions import col
df.filter((col('Item_Type')=='Soft Drinks') & (col('Item_Weight')<10)).display()
```

####scenario-3

```python
from pyspark.sql.functions import col
df.filter((col('Outlet_Size').isNull()) & (col('Outlet_Location_Type').isin('Tier 1','Tier 2'))).display()
```

### WITH COLUMN RENAMED

```python
df.withColumnRenamed('Item_Weight','Item_Wt').display()
```

###WITHCOLUMN

#### scenario-1 creating new column 

```python
from pyspark.sql.functions import lit
df = df.withColumn('flag',lit('new')).display()
```

##### while creating a new column we have to save it in a dataframe

####scenario-2 modifing the existing column

```python
from pyspark.sql.functions import col
df.withColumn('multiply',col('Item_Weight')*col('Item_MRP')).display()
```

####scenario-3 replacing the values 

```python
from pyspark.sql.functions import col
from pyspark.sql.functions import regexp_replace
df.withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Regular','Reg')).\
withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Low Fat','LF')).display()
```

###TYPE CASTING

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType
df = df.withColumn('Item_Weight',col('Item_Weight').cast(StringType()))
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType
df = df.withColumn('Item_Weight',col('Item_Weight').cast(StringType()))
```

### SORT/ORDER BY 

#### scenario-1 sorting based on single column using desc

```python
from pyspark.sql.functions import col
df.sort(col('Item_Weight').desc()).display()
```

#### scenario-2
#### sorting based on single column using asc

```python
from pyspark.sql.functions import col
df.sort(col('Item_Visibility').asc()).display()
```

#### scenario-3 
#### sorting based on multiple columns 

```python
from pyspark.sql.functions import col
df.sort(['Item_Weight','Item_Visibility'],ascending =[1,1]).display()
```

#### scenario-4
#### sorting based on multiple columns with both asc & desc

```python
from pyspark.sql.functions import col
df.sort(['Item_Weight','Item_Visibility'],ascending =[1,0]).display()
```
