###UNION AND UNION ALL

```python
df1.display()
```

```python
df2.display()
```

### DATE FUNCTIONS

#### Current Date 

```python
from pyspark.sql.functions import current_date
df = df.withColumn('Curr_Date',current_date())
```

```python
df.limit(5).display()
```

#### Date_Add

```python
from pyspark.sql.functions import date_add
df = df.withColumn('week_after',date_add(current_date(),7))
df.limit(5).display()
```

#### Date_Sub

```python
from pyspark.sql.functions import date_sub
df = df.withColumn('week_before',date_sub(current_date(),7))
df.limit(5).display()
```

```python
from pyspark.sql.functions import date_add
df = df.withColumn('week_before',date_sub(current_date(),-7))
df.limit(5).display()
```

####DateDIFF 

```python
from pyspark.sql.functions import datediff
df = df.withColumn('datediff',datediff('week_after','Curr_Date'))
df.limit(5).display()
```

#### DateFormat

```python
from pyspark.sql.functions import date_format
df = df.withColumn('week_before',date_format('week_before','dd-MM-yyyy'))
df.limit(5).display()
```

### Handling Nulls

#### Dropping Nulls


```python
df.dropna('all').display()
```

```python
df.dropna('any').display()
```

```python
df.dropna(subset=['Outlet_Size']).display()
```

#### Here the difference between 'all' & 'any' 
#### any-Drop the row if ANY column contains null.
#### all-Drop the row only if ALL columns are null.

#### Filling Nulls 

```python
df.fillna('NotAvailable',subset=['Outlet_Size']).display()
```

```python
df.fillna('NotAvailable').display()
```

### SPLIT AND INDEXING

####SPLIT

```python
from pyspark.sql.functions import split
df.withColumn('Outlet_Type',split('Outlet_Type',' ')).display()
```

####INDEXING

```python
from pyspark.sql.functions import split
df.withColumn('Outlet_Type',split('Outlet_Type',' ')[1]).limit(5).display()
```

#### EXPLODE


```python

df_exp = df.withColumn('Outlet_Type',split('Outlet_Type',' '))

df_exp.limit(5).display()
```

```python
from pyspark.sql.functions import explode
df_exp.withColumn('Outlet_Type',explode('Outlet_Type')).limit(2).display()
```

```python
from pyspark.sql.functions import array_contains
df_exp.withColumn('Type1_flag',array_contains('Outlet_Type','Type1')).limit(2).display()
```

### GROUPBY

#### scenario-1 

```python
from pyspark.sql.functions import sum
df.groupBy('Item_Type').agg(sum('Item_MRP')).display()
```

#### scenario-2

```python
from pyspark.sql.functions import avg
df.groupBy('Item_Type').agg(avg('Item_MRP')).display()
```

#### scenario-3

```python
from pyspark.sql.functions import sum
df.groupBy('Item_Type','Outlet_Size').agg(sum('Item_MRP').alias('Total_MRP')).display()
```

#### scenario-4 

```python
from pyspark.sql.functions import sum
from pyspark.sql.functions import avg
df.groupBy('Item_Type','Outlet_Size').agg(sum('Item_MRP'),avg('Item_MRP')).display()
```
